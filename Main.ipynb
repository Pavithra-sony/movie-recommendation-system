```python
movies = movies.reset_index(drop=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english') #Converts text (tags) into a numerical bag of words matrix
```

```python
vector = cv.fit_transform(movies['tags']).toarray() #Each movie’s tags become a vector of length 5000 (presence/absence of words)
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity #Measures closeness between two movies by angle between their tag-vectors
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
print(movies['title'].str.contains('Lego', case=False).sum())
print(movies[movies['title'].str.contains('Lego', case=False)])
```

```python

def recommend(movie):
    movie = movie.lower()
    movies['title_lower'] = movies['title'].str.lower()

    if movie not in movies['title_lower'].values:
        print(f"Movie '{movie}' not found in the dataset.")
        return

    index = movies[movies['title_lower'] == movie].index[0]     # Get index of the given movie
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1]) # Sort movies based on similarity score (highest first)
    
    print(f"\nTop 5 recommendations for '{movie.title()}':")
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
